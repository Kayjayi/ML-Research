# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
X= dataset.iloc[:,:-1].values # excludes the index of the last column
y = dataset.iloc[:,-1].values
```

```
print(X)
```

```
print(X)
```

***Importing the Dataset -Kay***


## Taking care of missing data

```
from sklearn.impute import SimpleImputer # works for numerical values
imputer = SimpleImputer(missing_values = np.nan, strategy='mean')
imputer.fit(X[:, 1:3]) # fit expects values with numerical values
X[:, 1:3] = imputer.transform(X[:,1:3]) # replace the new missing values with the mean
```

```
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])] , remainder='passthrough')
X = np.array(ct.fit_transform(X)) # it is important to change to numpy
```

```
print(X) # replaces the categorical data in the independent dataset
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,3:] = sc.fit_transform(X_train[:,3:]) # fit gets the mean and standard deviation. Transform conve
# convert the values. 
X_test[:,3:] = sc.transform(X_test[:,3:]) # apply only transform to the test set using the fit function
# used for X_train
```
